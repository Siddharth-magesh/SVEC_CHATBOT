```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load your dataset
data = pd.read_csv('crop_production_data.csv')

# Extract unique crop types
crop_types = data['Crop'].unique()

# Initialize an empty dictionary to store one-hot encoded values
one_hot_dict = {}

# One-hot encode each crop type and store in the dictionary
for crop_type in crop_types:
    # One-hot encode the crop type
    onehot_encoder = OneHotEncoder(sparse=False)
    # Extract rows where the crop type matches the current crop type
    temp_data = data[data['Crop'] == crop_type]
    # Fit and transform the one-hot encoder on the 'Crop' column of the filtered data
    onehot_encoded = onehot_encoder.fit_transform(temp_data[['Crop']])
    # Store the one-hot encoded values in the dictionary
    one_hot_dict[crop_type] = onehot_encoded.tolist()

# Print the dictionary
print(one_hot_dict)
```
